```python
train["target"].plot(kind="hist",figsize=(5,5))
plt.xlabel("loyalty_score")
plt.ylabel("card_index")
plt.title("loyalty score distribution")
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plt.rc('axes', titlesize=15)
plt.show()
```

```python
#the outlier list
q1=0.88
q3=0.77
_1_5iqr=(0.77+0.88)*1.5
lower_bound=q1-_1_5iqr
higher_bound=q3+_1_5iqr
```

```python
outlier_lst_L=train[["card_id","target"]][train["target"]<-1.59] 
outlier_lst_H=train[["card_id","target"]][train["target"]>3.25] 
norm_lst=pd.concat([train[["card_id","target"]][train["target"]>=-1.59],train[["card_id","target"]][train["target"]<=3.25]],axis=0)

all_card_loyal=train["target"].describe()
L_outlier_lst_card_loyal=outlier_lst_L.describe()
H_outlier_lst_card_loyal=outlier_lst_H.describe()
norm_lst_card_loyal=norm_lst.describe()

loyalscore_distribution_tb=pd.concat([all_card_loyal,pd.concat([pd.concat([norm_lst_card_loyal,L_outlier_lst_card_loyal],axis=1),H_outlier_lst_card_loyal],axis=1)],axis=1)

loyalscore_distribution_tb.columns=["All","-1.59=<majority<=3.25","outlier_lower_bound \n<-1.59 ","outlier_higher_bound \n>3.25"]
loyalscore_distribution_tb
```
